方法一：创建新节点

- 定义一个进位变量 `carry`，初始值为 `0`。
- 每次从两个链表中各取一个节点值：如果当前链表没有节点了，就补 `0`
- 然后计算：`sum = x + y + carry`
- 当前位应该存储：`sum % 10`，新的进位：`sum // 10`
- 为了方便构造结果链表，使用一个 **虚拟头节点 dummy**。

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        dummy = ListNode(0)   # 虚拟头节点
        cur = dummy
        carry = 0            # 进位

        while l1 or l2 or carry:
            x = l1.val if l1 else 0
            y = l2.val if l2 else 0

            total = x + y + carry
            carry = total // 10

            cur.next = ListNode(total % 10)
            cur = cur.next

            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next
```

```python
def build_linked_list(nums):
    dummy = ListNode()
    cur = dummy
    for num in nums:
        cur.next = ListNode(num)
        cur = cur.next
    return dummy.next

def linked_list_to_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result
```

```python
l1 = build_linked_list([2, 4, 3])
l2 = build_linked_list([5, 6, 4])

sol = Solution()
ans = sol.addTwoNumbers(l1, l2)

print(linked_list_to_list(ans))
```

# 方法二：哈希表优化

我们换一个思路。

对于当前数字 `num`，如果它想和另一个数加起来等于 `target`，那么我们只需要找：

`rest = target - num`

如果这个 `rest` 之前已经出现过，那么答案就找到了。

```python
class Solution(object):
    def twoSum(self, nums, target):
        hashmap = {}

        for i, num in enumerate(nums):
            rest = target - num

            if rest in hashmap:
                return [hashmap[rest], i]

            hashmap[num] = i

        return []
```

```python
nums = [2, 7, 11, 15]
target = 9

sol = Solution()
print(sol.twoSum(nums, target))
```
