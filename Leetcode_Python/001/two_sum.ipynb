
# 1. 两数之和（Two Sum）

给定一个整数数组 `nums` 和一个目标值 `target`，请你在该数组中找出和为目标值 `target` 的那两个整数，并返回它们的数组下标。

你可以假设：

- 每种输入只会对应一个答案
- 同一个元素不能使用两次
- 返回下标的顺序可以任意

---

方法一：暴力解法

最直接的想法是：

- 第一个数依次取 `nums[0]`、`nums[1]`、`nums[2]` ...
- 对于每个数，再向后找另一个数
- 如果两数之和等于 `target`，就返回它们的下标

```python
class Solution(object):
    def twoSum(self, nums, target):
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    return [i, j]
```

```python
# 测试
nums = [2, 7, 11, 15]
target = 9

sol = Solution()
print(sol.twoSum(nums, target))
```

# 方法二：哈希表优化

我们换一个思路。

对于当前数字 `num`，如果它想和另一个数加起来等于 `target`，那么我们只需要找：

`rest = target - num`

如果这个 `rest` 之前已经出现过，那么答案就找到了。

```python
class Solution(object):
    def twoSum(self, nums, target):
        hashmap = {}

        for i, num in enumerate(nums):
            rest = target - num

            if rest in hashmap:
                return [hashmap[rest], i]

            hashmap[num] = i

        return []
```

```python
nums = [2, 7, 11, 15]
target = 9

sol = Solution()
print(sol.twoSum(nums, target))
```
